```python
import os
import sys
print(sys.executable)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

import tensorflow as tf
import xgboost as xgb
import os
import warnings
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, TimeDistributed, Flatten, Dropout, RepeatVector
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller, kpss, ccf
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from math import sqrt
```

```python

energy_df = pd.read_csv("energy_dataset.csv")
weather_df = pd.read_csv("weather_features.csv")
```

```python
# convert time columns to UTC
energy_df['time'] = pd.to_datetime(energy_df['time'], utc=True)
weather_df['time'] = pd.to_datetime(weather_df['dt_iso'], utc=True)

weather_df = weather_df.drop(columns=['dt_iso'])
```

```python
# clean erngy dataset
# Drop unusable columns
energy_df = energy_df.drop([
    'generation fossil coal-derived gas','generation fossil oil shale', 
    'generation fossil peat', 'generation geothermal', 
    'generation hydro pumped storage aggregated', 'generation marine', 
    'generation wind offshore', 'forecast wind offshore eday ahead',
    'total load forecast', 'forecast solar day ahead',
    'forecast wind onshore day ahead'
], axis=1)

# Interpolate missing values
energy_df.interpolate(method='linear', limit_direction='forward', inplace=True)

# Set index
energy_df = energy_df.set_index('time')
```

```python
# clean weather data
# Convert int → float
def df_convert_dtypes(df, convert_from, convert_to):
    cols = df.select_dtypes(include=[convert_from]).columns
    for col in cols:
        df[col] = df[col].astype(convert_to)
    return df

weather_df = df_convert_dtypes(weather_df, np.int64, np.float64)

# Set index
weather_df = weather_df.set_index('time')
```

```python
# remove duplicates 
weather_df = weather_df.reset_index().drop_duplicates(
    subset=['time', 'city_name'],
    keep='first'
).set_index('time')
```

```python
# drop useless categories
weather_df = weather_df.drop([
    'weather_main', 
    'weather_description', 
    'weather_icon'
], axis=1)
```

```python
# Pressure outliers
weather_df.loc[weather_df.pressure > 1051, 'pressure'] = np.nan
weather_df.loc[weather_df.pressure < 931, 'pressure'] = np.nan

# Wind speed outliers
weather_df.loc[weather_df.wind_speed > 50, 'wind_speed'] = np.nan
```

```python
# group by city

# 1.3. Merging the two datasets

# Clean city names FIRST (important fix)
weather_df['city_name'] = weather_df['city_name'].str.strip()

# Split the df_weather into 5 dataframes (one for each city)
dfs = [x for _, x in weather_df.groupby('city_name')]

# Start from energy dataset
df_final = energy_df.copy()

# Merge each city's weather data
for df in dfs:
    city = df['city_name'].iloc[0]  # safe single value
    
    # remove city column
    df = df.drop(columns=['city_name'])
    
    # add city suffix to all weather columns
    df = df.add_suffix(f'_{city}')
    
    # merge on time index
    df_final = df_final.merge(
        df,
        left_index=True,
        right_index=True,
        how='outer'
    )

```

```python
print("Shape:", df_final.shape)
print("Missing values:", df_final.isnull().sum().sum())
print("Duplicates:", df_final.duplicated().sum())
```

```python
# plot time series 
def plot_series(df=None, column=None, series=pd.Series([]), 
                label=None, ylabel=None, title=None, start=0, end=None):
    sns.set()
    fig, ax = plt.subplots(figsize=(20, 6))
    ax.set_xlabel('Time')
    
    if column:
        ax.plot(df[column][start:end], label=label)
        ax.set_ylabel(ylabel)
        
    if series.any():
        ax.plot(series, label=label)
        ax.set_ylabel(ylabel)
        
    if label:
        ax.legend()
    if title:
        ax.set_title(title)
        
    ax.grid(True)
    return ax


plot_series(
    df=df_final,
    column='price actual',
    ylabel='Electricity Price (€/MWh)',
    title='Electricity Price (First 2 Weeks)',
    end=24*7*2
)

plt.show()
```

```python
# Plot the 'rain_1h' for Bilbao

ax = plot_series(df_final, 'rain_1h_Bilbao', 
                 label='Hourly', ylabel='Actual Price (€/MWh)',
                 title='Rain in the last hour in Bilbao')
plt.show()

# Plot the 'rain_3h' for Bilbao

ax = plot_series(df_final, 'rain_3h_Bilbao', 
                 label='Hourly', ylabel='Rain in last 3 hours (mm)',
                 title='Rain in the last 3 hours in Bilbao')
plt.show()
```

```python
# drop 3h

df_final = df_final.drop(
    columns=[col for col in df_final.columns if 'rain_3h' in col]
)
```

```python
# Plot the hourly actual electricity price, along with the weekly rolling mean

rolling = df_final['price actual'].rolling(24*7, center=True).mean()
ax = plot_series(df_final, 'price actual', label='Hourly', ylabel='Actual Price (€/MWh)',
                 title='Actual Hourly Electricity Price and Weekly rolling mean')
ax.plot(rolling, linestyle='-', linewidth=2, label='Weekly rolling mean')
plt.show()
```
